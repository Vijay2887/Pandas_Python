```python
# Required Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college_df = pd.read_excel("college_data.xls")
top_students_df = college_df[college_df["FINAL_CGPA"] >= 9 ].sort_values(by="FINAL_CGPA", ascending=False).copy().reset_index()
top_students_df.drop("index", axis=1, inplace=True)
top_students_df
```

```python
# Number of males in the top 12 with cgps > 9
top_male_students_df = top_students_df[top_students_df["GENDER"] == "Male"]
top_male_students_df
top_male_students_df.pivot_table(index="HTNO", values="FINAL_CGPA").plot(kind="line")
```
